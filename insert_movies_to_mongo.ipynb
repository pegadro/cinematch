```python
import pandas as pd
```

```python
movies_10000_full = pd.read_csv("./datasets/custom-ml/movies_10000_full.csv")
movies_10000_full = movies_10000_full.rename(columns={"Unnamed: 0": "index"})
movies_10000_full.head()
```

```python
movies_10000_full_dict = movies_10000_full.to_dict("records")
movies_10000_full_dict
```

```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client['cinematch']
collection = db['movies']
```

```python
collection.insert_many(movies_10000_full_dict)
```
