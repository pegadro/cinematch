```python
import pandas as pd
import numpy as np
```

```python
movies_10000 = pd.read_csv("./datasets/custom-ml/movies_10000.csv")
movies_10000.head()
```

```python
movies_posters_10000 = pd.read_csv("./datasets/custom-ml/movies_posters_10000.csv")
movies_posters_10000.head()
```

```python
movies_posters_10000["poster_url"][movies_posters_10000["poster_url"].isna()]
```

```python
movies_10000_full = pd.merge(movies_10000, movies_posters_10000[["movieId", "tmdbId", "poster_url"]], on="movieId", how="inner")
movies_10000_full.head()
```

```python
movies_10000_full[movies_10000_full.columns.difference(["ratings", "mean_rating"])].to_csv("./datasets/custom-ml/movies_10000_full.csv", index=True)
```
