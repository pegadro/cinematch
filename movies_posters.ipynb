```python
import pandas as pd
import numpy as np
```

```python
movies_10000 = pd.read_csv("./datasets/custom-ml/movies_10000.csv")
movies_10000.head()
```

```python
moviesIds_10000 = movies_10000["movieId"]
moviesIds_10000
```

```python
links_full = pd.read_csv("./datasets/ml-latest/links.csv")
links_full.head()
```

```python
tmbids = links_full[links_full["movieId"].isin(moviesIds_10000)]
tmbids
```

```python
links_10000 = links_full[links_full["movieId"].isin(moviesIds_10000)]
```

```python
merged_movies_10000 = pd.merge(movies_10000, links_10000, on="movieId", how="inner")
merged_movies_10000.head()
```

```python
merged_movies_10000.info()
```

```python
import requests
import json

API_KEY = "bb782d9682f1bc5cd45abf0151831dc2"

posters = []

for i in range(len(merged_movies_10000)):
    dict_movie_poster = {}
    
    movie = merged_movies_10000.iloc[i]
    movie_tmdbid = movie["tmdbId"]
    
    url = f"https://api.themoviedb.org/3/movie/{movie_tmdbid}?api_key={API_KEY}"
    
    response = requests.get(url)
    response_dict = json.loads(response.text)
    
    poster_path = ""
    
    try:
        poster_path = response_dict["poster_path"]
    except KeyError:
        poster_path = None
    
    poster_url = f"https://image.tmdb.org/t/p/original{poster_path}" if poster_path != None else None
    
    dict_movie_poster["movieId"] = movie["movieId"]
    dict_movie_poster["tmdbId"] = movie_tmdbid
    dict_movie_poster["title"] = movie["title"]
    dict_movie_poster["poster_url"] = poster_url
    
    posters.append(dict_movie_poster)
    
    if i % 100 == 0:
        print(i)
    
    #print(movie["title"], " - ", poster_url)
```

```python
posters
```

```python
posters_df = pd.DataFrame(posters)
posters_df.sample(10)
```

```python
posters_df.to_csv("./datasets/custom-ml/movies_posters_10000.csv", index=False)
```
