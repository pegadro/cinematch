```python
import sys
import os

current_dir = os.getcwd()
parent_dir = os.path.abspath(os.path.join(current_dir, '..'))
if parent_dir not in sys.path:
    sys.path.insert(0, parent_dir)

from db import db
from utils.recommender import RatingsNormalizer, get_ratings_binary_matrix
```

```python
ratings_matrix = db.load_sparse_ratings("cinematch_regression").toarray()
```

```python
ratings_normalizer = RatingsNormalizer()
ratings_normalizer.fit(ratings_matrix, R=get_ratings_binary_matrix(ratings_matrix))
```

```python
ratings_normalizer.means_
```

```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
db = client["cinematch_regression"]
collection_mean_movies_rating = db["mean_rating"]
```

```python
collection_mean_movies_rating.insert_one({"means":ratings_normalizer.means_.tolist()})
```

```python
mean_rating = []

for i in range(len(ratings_normalizer.means_)):
    mean_rating.append({
        "index": i,
        "mean_rating": ratings_normalizer.means_[i]
    })
```

```python
collection_mean_movies_rating.insert_many(mean_rating)
```
