```python
import numpy as np
```

```python
movies_latent_factors = np.load("../datasets/custom-ml/regression/movies_latent_factors.npz")["data"]
users_latent_factors = np.load("../datasets/custom-ml/regression/users_latent_factors.npz")["data"]
intercept_latent_factors = np.load("../datasets/custom-ml/regression/intercept_latent_factors.npz")["data"]
```

```python
intercept_latent_factors_list = []

for i in range(len(intercept_latent_factors[0])):
    intercept_latent_factors_list.append({
        "index": i,
        "data": intercept_latent_factors[0][i]
    })
```

```python
users_latent_factors_list = []

for i in range(users_latent_factors.shape[0]):
    users_latent_factors_list.append({
        "index": i,
        "data": users_latent_factors[i].tolist()
    })
```

```python
movies_latent_factors_list = []

for i in range(movies_latent_factors.shape[0]):
    movies_latent_factors_list.append({
        "index": i,
        "data": movies_latent_factors[i].tolist()
    })
```

```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
db = client["cinematch_regression"]
collection_intercept_latent_factors = db["intercept_latent_factors"]
collection_users_latent_factors = db["users_latent_factors"]
collection_movies_latent_factors = db["movies_latent_factors"]
```

```python
collection_intercept_latent_factors.insert_many(intercept_latent_factors_list)
```

```python
collection_users_latent_factors.insert_many(users_latent_factors_list)
```

```python
collection_movies_latent_factors.insert_many(movies_latent_factors_list)
```
